# MarkUs Guide

Each week, you will submit notebooks from JupyterHub to MarkUs for grading.

## Step 1: Download this notebook
1. Download this Jupyter notebook file to your computer: `File —> Download as —> Notebook (.ipynb)`.

<center><img src='download.png' width="544"></center>

### WARNING: if you download twice, you need to rename the file

If you download a second time and the file already exists locally on your computer, you'll get a new file with a number appended. (For example: `CJH332_tutorial1-1.ipynb` is used the second time you download, then `CJH332_tutorial1-2.ipynb`, and so on.) Before you submit, you need to:

1. Delete the old notebook download.
2. Rename your most current download without the appended number. (For example: Change `CJH332_tutorial1-1.ipynb` to `CJH332_tutorial1.ipynb`.

## Task 2: Upload to MarkUs

1. Log in here: https://markus.teach.cs.toronto.edu/markus/ (Tip: Control/Command-click to open it in a new tab so you can still see these instructions.)
2. Choose your course. (For example: `CJH332-2026-01`)
3. Click the assessment for your submission. (For example: `tutorial1`)
4. Click the `Submissions` tab.
5. Click button `Submit File` on the bottom right.
6. Click button `Choose Files`. (Rename the filename (with the extensions), if needed, to match the Required files name.)
    <center><img src='Required_files.png' width="434"></center>
7. Select the file that you wish to submit, then click `Save`. (You should see a checkmark if the required files are uploaded).
<center><img src = 'Success.png' width="359"></center>

**Note**: If you would like to make changes to a submission, first select the file and delete it from MarkUs, then repeat steps 5 through step 7.

## Task 3: Run the automated tests

Once you have uploaded your submission to MarkUs, there are automated tests that check if the desired outputs and variables are defined properly in your notebook, and you can (and should) run the tests. 

1. Click the `Automated Testing` tab.
2. Click button `Run tests` and wait several seconds while the test is run.
3. If you don't see the test results, reload the webpage.
