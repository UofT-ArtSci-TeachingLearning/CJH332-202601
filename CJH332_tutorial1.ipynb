### Variables

A variable is just a name you give to a piece of information. It’s like putting a label on a box so you know what’s inside.

In Python, variable names can use letters, numbers, and underscores, but they can’t start with a number. They’re also case-sensitive, so age and Age would count as two different variables.

You create a variable using the `=` sign.
Example: my_age = 25

After you make a variable, you can use it anywhere in your code— print it, do math with it, save results, and more.

### Data types
Every value in Python has a *data type*, which determines how the computer stores it and what operations you can perform. For example, an int (integer) represents whole numbers and supports arithmetic, while a float holds decimal values. Other common types include strings for text and booleans for true/false logic. 

 Here are a few basic types you will encounter:

 | Type | Description | Example |
 |------|-------------|---------|
 | **int** | Whole numbers | 10, -3, 42 |
 | **float** | Decimal numbers | 3.14, 0.01 |
 | **str** | Text strings | "Hello", "Python" |
 | **bool** | True/False values | True, False |

```python
x = 10
x
```

```python
name = 'Julia'
name
```

### Comments

Comments let you add notes or explanations to your code. They’re meant for humans, not the computer! In Python, a comment starts with a # symbol. Everything after the # on that line is ignored by Python when the code runs. This makes comments a great way to explain what your code is doing or to leave reminders for yourself.

```python
# This is a comment. It is not evaluated when we run our code
```

### Operators
Python supports several categories of operators. Some basic ones include

 | Operator | Meaning | Example | Result |
 |----------|---------|---------|--------|
 | `+` | Addition | `5 + 2` | 7 |
 | `-` | Subtraction | `5 - 2` | 3 |
 | `*` | Multiplication | `5 * 2` | 10 |
 | `/` | Division | `5 / 2` | 2.5 |
 | `**` | Exponentiation | `5 ** 2` | 25 |

### Built-in functions
Python includes many useful functions right out of the box. A function is like a small tool that performs a specific task. For example, to show something on the screen, you can use the print() function.

```python
a = 5
b = 2

print(a + b) # Addition
print(a - b) # Subtraction
print(a * b) # Multiplication
print(a / b) # Division
print(a ** b) # Exponentiation
```

Python includes many more helpful built-in functions you can use right away. For example:

`abs()` gives the absolute value of a number

`round()` rounds a number to the nearest whole number (or a chosen number of decimal places)

```python
abs(-100) 
```

```python
round(2.75)
```
