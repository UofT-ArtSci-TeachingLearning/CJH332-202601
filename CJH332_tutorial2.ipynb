### Collections of values

So far, we’ve only worked with single values like integers, floats, strings. But in real programs, we often need to work with groups of values — such as a list of numbers, names, or measurements.

Python provides built-in data types that let us store multiple values together. One of the most common is a list.

A list lets you keep several items in one place and work with them as a group. You create a list by putting values inside square brackets `[]`, separated by commas.

```python
my_grades = [90,88,70]
```

### Accessing items in a list

Lists are ordered, meaning each item has a specific position called an index. This lets you access individual items using square brackets.

You can use these indexes to get a single item from the list.

```python
my_grades[0] #returns the first grade in my list
```

**Slice** a list to get a range of items

```python
my_grades[0:2] #returns the first two grades in my list; items from index 1 up to (but not including) index 2
```

### Adding items to a list

Python provides a few different functions to add items to a list: append() will add items to the end of a list, while insert() will insert an item at a specific location.

```python
temperatures = [10,33,55]
temperatures.append(15)
temperatures
```

```python
temperatures.insert(1,15) # insert at index 1
temperatures
```

With the insert() function the first value is the index position that you wish to insert the value to. Note with Python, like many programming languages, index values start at 0. So the first position in the list has the index of 0

### Removing items from a list

Similarly Python provides a few functions to remove items from a list, including del, pop() and remove()

The remove function removes the first instance of an item found in a list

```python
my_list = [1,2,3,4,3,5]
my_list.remove(3)
my_list
```

The del keyword removes an item from a list at a specific index

```python
del my_list[3]
my_list
```

Finally, the pop function removes an item from a specific index from the list and allows it to be assigned to a variable.

```python
former_last_item = my_list.pop(3)
former_last_item
```

```python
my_list
```

### Length of lists

It can be helpful to know the length of a list. The function len() helps us do that

```python
the_list = [1,34,22,11,10]
list_length = len(the_list)
list_length
```

### Lists can be more than just numbers

Lists in Python can contain many Python data types as items, including multiple data types in one list. A list can even contain another list. The following are all valid examples of lists:  

```python
list_example1 = [1,2.3,"Toronto","Ontario"]
list_example2 = [2,[3,4,5],6,"Bob"]
list_example3 = ["English", "French", "Spanish"]
```
